```python
# Importar la librería de Python que maneja fracciones 
from fractions import Fraction as fr
 
#La función bernoulli pide como parámetro el índice n , para el número de coeficientes que queremos obtener
def bernoulli(n):
    #El coeficiente lo representa A.
    A = [0] * (n+1)
    
    for m in range(n+1):
        A[m] = fr(1, m+1)
        for j in range(m, 0, -1):
          A[j-1] = j*(A[j-1] - A[j])
    return A[0] # (El cual es Bn. Los n impares anulan elcoeficiente por eso no se toman en cuenta)
 
bn = [(i, bernoulli(i)) for i in range(101)] #Se pone 101 dado que el for empieza en 0.
bn = [(i, b) for i,b in bn if b]
width = max(len(str(b.numerator)) for i,b in bn)
for i,b in bn:
    print('B(%2i) = %*i/%i' % (i, width, b.numerator, b.denominator))
```
